```python
class Cliente:
    def __init__(self,nombre,apellido,dni):
        self.nombre=nombre
        self.apellido=apellido
        self.dni=dni
    def __str__(self):
        return(f"Nombre: {self.nombre}\n"
               f"Apellido: {self.apellido}\n"
               f"DNI: {self.dni}\n")
class Banco:
    def __init__(self):
        self.usuario=[]
    def agregar_cliente(self):
        nombre=input("\n Nombre del cliente:")
        apellido=input("\n Apellido del cliente:")
        dni=int(input("\n DNI del cliente:"))
        
        if self.buscar_por_dni(dni):
            print("Ya existe un usuario con el mismo DNI")
        else:
            self.usuario.append(Cliente(nombre,apellido,dni))
            print("El cliente se agrego con exito")
                
    def buscar_por_dni(self,dni):
        for usuario in self.usuario:
                if usuario.dni==dni:
                    return usuario   
    
    def mostrar_cliente(self,dni):
        cliente=self.buscar_por_dni(dni)
        if cliente:
            print("\n" , cliente)
        else:
            print("\n No existe un cliente con ese DNI")

def main():
    banco=Banco()
    print("\nBanco Nacional del Perú")
    
    while True:
        print("\nDigite una opción : \n\n"
              " 1. Agregar Usuario\n"
              " 2. Visualizar usuario\n"
              " 3. Salir")
        opcion=input("\n Digite una opcion: ")
        
        if opcion =="1":
            banco.agregar_cliente()
        elif opcion=="2":
            dni=int(input("Digite el DNI del cliente:"))
            banco.mostrar_cliente(dni)
        elif opcion=="3":
            print("Adios del sistema del banco")
            break
        else:
            print("\nOpción inválida, intentalo nuevamente")
main()
```
