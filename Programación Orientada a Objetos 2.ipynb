## Composición
- Es un concepto que nos permite modelar la relacion has-a (tiene un ...) objeto entre clases.
Significa que la clase tiene un componente(objeto), desde el punto de vistas del programador implica insertar objetos en otro objeto.
- A diferencia de la Herencia tiene una relacion is-a(es un tipo de ...)por ejemplo una calse hija puede utilizar atributos y metodos de la clase padre

![Image](Composicionsobrelaherencia.png)

```python
class Casa:
    def __init__(self,call,num,ciu,p):
        self.calle=call
        self.numero=num
        self.ciudad=ciu
        self.pais=p
        self.recamara=Cuarto('recamara',5,5) #ESTOY REALIZANDO LA COMPOSICION
        self.cocina=Cuarto('cocina',6,4) #ESTOY REALIZANDO LA COMPOSICION
        self.baño=Cuarto('baño',3,4) #ESTOY REALIZANDO LA COMPOSICION
        self.sala=Cuarto('sala',7,6) #ESTOY REALIZANDO LA COMPOSICION
        
    def dimensioncasa(self):
        total=0
        #ESTO SE ENCARGARA DE TOMAR VALORES DE CADA CUARTO(EL LARGO Y EL ANCHO)
        for c in (self.recamara,self.cocina,self.baño,self.sala):
            total=total+c.tamaño()
        return total
    def direccion(self):
        print('\nDireccion: ')
        print(self.calle+''+self.numero)
        print(self.ciudad+','+self.pais+'\n')

class Cuarto:
    def __init__(self,d,l,a):
        self.descripcion=d
        self.largo=l
        self.ancho=a
    def tamaño(self):
        return self.largo*self.ancho
    
micasa=Casa('Doña ANA','204','Lima','Perú')
print('En total la casa tiene ' +str(micasa.dimensioncasa())+' metros cuadrados')
micasa.direccion()
```

### Encapsulamiento
- Es la técnica de ocultar datos que me permite que mis atributos de un objeto no puedan ser accedidos desde fuera de la definición  de la clase. Para ello es necesario nombrar los atributos con un prefijo de doble subrayado __

```python
class Factura:
    #ATRIBUTO ENCAPSULADO O PRIVADO
    __igv=18
    def __init__(self,unidad,precio):
        self.unidad=unidad
        self.precio=precio
    def pagar(self):
        total=self.unidad*self.precio
        impuesto=total*Factura.__igv/100
        return(total+impuesto)

laptop = Factura(1,1700)
print(laptop.pagar(), " soles")

#PRODUCIR ERROR PORQUE EL ATRIBUTO ES PRIVADO
# print("El igv utilizado es: ",laptop__igv)
```

```python
class Factura_Accediendo:
    __igv=18
    def __metodo_privado(self):
        print("Este es un metodo privado de la misma clase")
    def atribtuo_publico(self):
        return self.__igv
    def metodo_publico(self):
        return self.__metodo_privado()

camara =Factura_Accediendo()
print(camara.atribtuo_publico())
camara.metodo_publico()
```

## Polimorfismo
- Se refiere a la propiedad por la que es posible enviar mensajes con el mismo nombre a objetos de tipos distintos. El único requisito que debe cumplir los objetos es saber responder el mensaje que se le envia

```python
class Gato:
    def ruido(self):
        print('Miau')

class Perro:
    def ruido(self):
        print('Guau,Guau')

class Humano:
    def ruido(self):
        print('Me gusta Python')
        
def emitesonido(objetoType):
    objetoType.ruido()  #TRES INSTANCIAS DIFERENTES LLAMAN A LA ACCION USANDO EL MISMO METODO

objGato = Gato()
objPerro= Perro()
objHumano=Humano()

emitesonido(objHumano)
```
